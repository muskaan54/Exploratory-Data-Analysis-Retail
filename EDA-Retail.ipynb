#### EXPLORATORY DATA ANALYSIS

```python
#drawing a pair plot
sns.pairplot(store)
```

```python
#HEAT MAP
store.corr()
plt.figure(figsize=(10,5))
sns.heatmap(store.corr(),annot=True)
plt.show()
```

##### REMARK- heat map shows that discount and profit are negatively correlated
##### profit and quantity are weakly correlated
##### sales and profit are moderately correlated

```python
#drawing a scatter plot
plt.scatter(store['Sales'],store['Profit'], color= 'orange')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales vs Profit ')
plt.show()
```

##### REMARK- when sales are high, company mostly gains profit
##### when sales are low company may incur profit or loss

```python
sns.countplot(x=store['Ship Mode'])
```

##### REMARK- for majority of the people, favorable ship mode is standard class

```python
sns.countplot(x=store['Segment'])
```

```python
sns.countplot(x=store['Category'])
```

```python
store.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['red','green'],figsize=(8,5))
plt.ylabel('Profit')
plt.show()
```

##### REMARK- technology is giving the company maximum profit followed by office supplies

```python
store.groupby('Sub-Category')[['Profit','Sales']].sum().plot.bar(color=['red','green'],figsize=(8,5))
plt.ylabel('Profit')
plt.show()
```

##### REMARK- copiers are providing maximum profit whereas tables are giving loss

```python
store['Sub-Category'].value_counts()
```

##### REMARK- binders and paper is the most ordered item whereas copiers are less ordered

```python
sns.barplot(x='Region', y ='Profit',data=store)
plt.show()
```

##### REMARK= this plot between profit and region show that profit was made more in west and less in central part

```python
fig, axes =plt.subplots(2,2,figsize= (16,8))
fig.suptitle("sales with different segments and shipping modes", fontsize=15)
sns.barplot(data= store, x='Ship Mode', y='Sales',ax=axes[0,0],dodge= False)
sns.lineplot(data =store,x='Ship Mode',y='Sales',color = 'green',ax=axes[0,1])
sns.barplot(data=store,x='Segment',y= 'Sales',ax=axes[1,0])
sns.lineplot(data = store,x='Segment',y='Sales',ax=axes[1,1])
plt.show()
```

##### REMARK- the mode of shipping doesnt affect much to the sales 
##### whereas considering segment, home office provides highest sales.

```python
figure=store['State'].value_counts()
figure.plot(kind='bar', figsize=(15,12))
plt.show()
```

##### REMARK - Highest buyers are from California followed by New York and Texas 

```python
store.groupby('State')[['Profit','Sales']].sum().plot.bar(color=['red','green'],figsize=(18,8))
plt.ylabel('Profit')
plt.show()
```

##### REMARK- selling a product in california and new york provides huge profits to the company

```python
sns.lineplot(data=store,x ='Discount',y='Profit',color='teal')
plt.show
```

 ##### REMARK - When discount hits 30% to 50 % company incurrs a huge loss
